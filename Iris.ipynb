```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

sns.set(style="whitegrid")

# Step 2: Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Step 3: Display first rows
df.head()
```

```python
# Check data types of each column
print("üîç Data Types:")
print(df.dtypes)

# Check for missing values
print("\nüßº Missing Values:")
print(df.isnull().sum())
```

```python
# Group by species and get average for each flower type
print("\nüìà Average measurements per species:")
print(df.groupby("species", observed=True).mean())
```

```python
# Group by species and get average for each flower type
print("\nüìà Average measurements per species:")
print(df.groupby("species", observed=True).mean())
```

```python
# Line chart: Petal length over entries (just simulating time)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["petal length (cm)"], color='purple')
plt.title("Petal Length Over Time (Sample Index)")
plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```

```python
# Manually calculate average sepal length per species to avoid warning
avg_sepal = df.groupby("species", observed=True)["sepal length (cm)"].mean().reset_index()

# Plot it
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_sepal, x="species", y="sepal length (cm)", palette="pastel")
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
# Histogram: Distribution of petal length
plt.figure(figsize=(8, 5))
plt.hist(df["petal length (cm)"], bins=20, color='orchid', edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot: Sepal length vs. Petal length
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="sepal length (cm)", y="petal length (cm)", hue="species", palette="deep")
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```

```python
# Observations and Findings

1. **Petal Length Trend**: The petal length seems to increase steadily across the dataset, showing a slight variation as we move through the entries.
   
2. **Average Sepal Length by Species**: 
   - `Setosa` species has the smallest average sepal length, followed by `Versicolor` and `Virginica`.
   - `Virginica` species tends to have the largest sepal length on average.

3. **Petal Length Distribution**:
   - Petal lengths are mostly concentrated between 1 and 6 cm, with some outliers.
   - The distribution looks roughly normal, but the higher end has fewer values.

4. **Sepal vs. Petal Length**:
   - There's a positive correlation between sepal length and petal length. `Setosa` flowers have the smallest petals, while `Versicolor` and `Virginica` have larger petals.
   - The scatter plot also shows that petal length is a good indicator for species differentiation.
ve correlation between sepal length and petal length. `Setosa` flowers have the smallest petals, while `Versicolor` and `Virginica` have larger petals. - The scatter plot also shows that petal length is a good indicator for species differentiation.
```
